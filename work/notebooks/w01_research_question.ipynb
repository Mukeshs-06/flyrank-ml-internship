## 2. The question: decision, action, cost of a wrong call

- **Decision improved:** Which declining or underperforming content items should an editorial team prioritize updating first to mitigate traffic loss?
- **Who acts and what they do:** Content editors and SEO strategists. They inspect prioritized pages, update outdated information, refresh keywords/metadata, or restructure content to restore search relevance.
- **Cost of a wrong call:**
  - *False Positive (flagging a healthy page):* Wastes ~3–5 hours of editor bandwidth and budget on unnecessary content rewrites.
  - *False Negative (missing a severely declining page):* Leads to compounding organic search traffic loss and reduced audience engagement before intervention occurs.
- **Why data/ML helps:** A simple rule (e.g., "refresh every page older than 12 months") ignores multi-dimensional search dynamics. Machine learning enables ranking pages by combining multiple non-linear signals (impressions, CTR trends, position shifts, and engagement rates) into an explainable priority score.

## 3. Quick look at the data (2-3 real numbers)

Below, we load the starter dataset (`data/raw/content_refresh_anonymized.csv`) and compute key metrics that illustrate the scale of content decay within this dataset.

```python
import pandas as pd

# Load starter dataset relative to work/notebooks/
df = pd.read_csv("../../data/raw/content_refresh_anonymized.csv")

# Calculate real numbers
total_pages = len(df)
declining_pages = (df["trend_direction"] == "down").sum()
pct_declining = (declining_pages / total_pages) * 100
total_clicks = df["clicks_90d"].sum()
declining_clicks = df[df["trend_direction"] == "down"]["clicks_90d"].sum()
pct_declining_clicks = (declining_clicks / total_clicks) * 100

print(f"Total content items evaluated: {total_pages:,}")
print(f"Declining content items (trend_direction == 'down'): {declining_pages:,} ({pct_declining:.2f}%)")
print(f"Total 90-day clicks across dataset: {total_clicks:,}")
print(f"Clicks from declining pages: {declining_clicks:,} ({pct_declining_clicks:.2f}%)")
```

## 4. Careful words: what I can and can't claim

- **What can be claimed:**
  - *Observed performance:* Within this dataset of 30,000 pages across 32 clients, 54.21% of pages are in a downward traffic trend.
  - *Directional patterns:* We can measure statistical correlations between feature changes (e.g., impression trends, average position shifts) and traffic decline.
  - *Decision-support utility:* We can produce a ranked recommendation queue to help editors prioritize content review.
- **What CANNOT be claimed:**
  - *Causal proof:* We cannot claim that updating a page will guarantee ranking recovery or specific click gains.
  - *Algorithm prediction:* We do not claim to predict or reverse-engineer Google's search algorithm.
